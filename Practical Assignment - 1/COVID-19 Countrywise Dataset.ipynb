```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covidData = pd.read_csv('country_wise_latest.csv')
rows = covidData.sample()
columns = []
columns = covidData.columns
print("Features of the Dataset:")
for i in columns:
    print(i)

print()
print("A sample from the dataset:")
print(rows.head())
```

```python
isMissing = False
for i in columns:
    if(covidData[i].isnull == True):
        isMissing = True
        break
if(isMissing):
    print("Missing values exist")
else:
    print("Missing Values does not exist")

for i in columns:
    if(covidData[i].isnull == True):
        covidData[i].fillna(covidData[i].mean(), inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.title("Double Bar Chart of COVID-19 cases in WHO region")

plt.ylim(0, 500000)
labels = ['Confirmed Cases', 'Deaths']

plt.bar(covidData['WHO Region'], covidData['Confirmed'], color='blue', width = 0.3)
plt.bar(covidData['WHO Region'], covidData['Deaths'], color='red', width = 0.3)
plt.legend(labels=labels)
```

```python
# Pie chart of confirmed, active, death and recovered cases
plt.figure(figsize=(16, 16))
plt.title('Bar Plot b/w confirmed cases and active cases')

sliceDivisions = []
sliceDivisions.append(covidData['Confirmed'].sum())
sliceDivisions.append(covidData['Active'].sum())
sliceDivisions.append(covidData['Deaths'].sum())
sliceDivisions.append(covidData['Recovered'].sum())

labels=['Confirmed', 'Active', 'Deaths', 'Recovered']

plt.pie(sliceDivisions, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.legend(labels, loc='upper right')
```
