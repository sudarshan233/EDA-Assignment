```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
irisData = load_iris()
rows = irisData.data
dataFrame = pd.DataFrame(rows, columns = irisData.feature_names)
dataFrame['species'] = irisData.target

# dataFrame['species'] = dataFrame['species'].map({0: 'Setosa',
#                                                  1: 'Versicolor',
#                                                  2: 'Virginica'})
# print(dataFrame.shape)
# print(dataFrame.head())
# print(dataFrame.describe)
list = dataFrame['species']
count = 0
for i in list:
    if(i == 0):
        count += 1
print(count)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data = dataFrame, x = "species", y = "sepal length (cm)")
plt.xlabel("Different types of Species")
plt.ylabel("Sepal length in cm")
plt.title("Box Plot of Sepal Length")
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data = dataFrame, x = 'species', y = 'petal length (cm)')
plt.title("Violin Plot of Petal Length")
plt.xlabel('Different types of species')
plt.ylabel('Petal length in cm')
```

```python
plt.figure(figsize=(10,5))
plt.hist(data = dataFrame, x = 'sepal width (cm)', bins = 10, color = 'skyblue', edgecolor = 'black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.pairplot(data = dataFrame, hue = 'species')
```

```python
plt.figure(figsize=(10,5))
x = dataFrame['sepal length (cm)']
y = dataFrame['sepal width (cm)']
plt.scatter(x, y)
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal width in cm')
```

```python
correlationMatrix = dataFrame.corr()
plt.figure(figsize=(10, 6))
plt.title("Correlation Heatmap between features")
sns.heatmap(correlationMatrix, annot = True, cmap = 'viridis')
```
